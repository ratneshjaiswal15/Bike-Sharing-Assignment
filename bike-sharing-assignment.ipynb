## Step-3: Data Preperation for Model

```python
## dropping off columns which are not required for analysis.
bike.drop(['instant','dteday','casual','registered'],axis = 1,inplace = True)
bike.head()
```

```python
# converting the numeric values to corresponding catregorical data
bike.weekday = bike.weekday.map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})
bike.weathersit = bike.weathersit.map({1:'Clear',2:'Mist & Cloudy', 
                                             3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
bike['mnth'] = bike['mnth'].apply(lambda x: calendar.month_abbr[x])
bike.season = bike.season.map({1: 'Spring',2:'Summer',3:'Fall',4:'Winter'})
```

```python
#verify the tranformed data
bike.head()
```

```python
#create dimmues var for given catergory data
category_data = bike[['season','mnth','weekday','weathersit']]
catg_dummeis_data = pd.get_dummies(category_data,drop_first=True )
```

```python
# verify the catg_dummeis_data
catg_dummeis_data.head()
```

```python
#drop old columns and add new transformed catg_dummeis_data.
bike.drop(['season', 'mnth', 'weekday','weathersit'], axis = 1, inplace = True)
bike = pd.concat([catg_dummeis_data,bike],axis = 1)
bike.head()
```

```python
#varify all the columns
bike.info()
```

```python
bike.shape
```

### Spliting data frame into tarin and test data set.

```python
#Spliting the data into tarin and test data set.
bike_train, bike_test = train_test_split(bike, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescalling the feature: numeric var

```python
### Rescalling the feature except dummies var
scaler = MinMaxScaler()
num_vars = ['cnt','hum','windspeed','temp','atemp']
bike_train[num_vars] = scaler.fit_transform(bike_train[num_vars])
bike_train.head()
```

```python
bike_train.describe()
```

```python
bike_train.describe()
```

```python
# analyze the cofficient
plt.figure(figsize = (20, 10))
sns.heatmap(bike_train.corr(), annot = True)
plt.show()
```

# Step- 4: Building the Model

```python
# divide into X and y
y_bike_train = bike_train.pop('cnt')
X_bike_train = bike_train
```

```python
# unsing RFE with the 13 output varibales
lr = LinearRegression()
lr.fit(X_bike_train, y_bike_train)

rfe = RFE(lr, 13)            
rfe = rfe.fit(X_bike_train, y_bike_train)
```

```python
#Columns selected by RFE and their weights
list(zip(X_bike_train.columns,rfe.support_,rfe.ranking_))
```

### Model- 1

```python
# columns supported by RFE
col = X_bike_train.columns[rfe.support_]
col
```

```python
# columns not supported by RFE
X_bike_train.columns[~rfe.support_]
```

```python
# Creating RFE selected variable and also adding consntant for stats model.
X_bike_train_rfe = X_bike_train[col]
X_bike_train_rfe = sm.add_constant(X_bike_train_rfe)
lm = sm.OLS(y_bike_train,X_bike_train_rfe).fit()
lm.params
```

```python
print(lm.summary())
```

```python
# dropping off since mnth_jan > .05
X_bike_train_new = X_bike_train_rfe.drop(["mnth_Jan"], axis = 1)
```

### Model-2

```python
# Adding a constant variable 
X_bike_train_lm = sm.add_constant(X_bike_train_new)
lm = sm.OLS(y_bike_train,X_bike_train_lm).fit() 
print(lm.summary())
```

```python
### identify the multileinerlity
# Calculate the VIFs for the new model
vif = pd.DataFrame()
X = X_bike_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# const vif value is greaet than 5, so droopng off.
X_bike_train_new = X_bike_train_new.drop(['const'], axis=1)
```

```python
#Caluclate the new vif
vif = pd.DataFrame()
X = X_bike_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Model 3- Finalized

```python
#dropping hum from the model
X_bike_train_new = X_bike_train_new.drop(['hum'], axis=1)
# adding a cons var 
X_bike_train_lm = sm.add_constant(X_bike_train_new)
lm = sm.OLS(y_bike_train,X_bike_train_lm).fit()   
print(lm.summary())
```

```python
# Calculate the VIFs
vif = pd.DataFrame()
X = X_bike_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step- 5: Residual Analysis

```python
y_bike_train_cnt = lm.predict(X_bike_train_lm)

# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_bike_train - y_bike_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)  
```

## Step- 6: Predictions using the finalized model.

```python
no_of_variables = ['cnt','hum','windspeed','temp','atemp']
bike_test[num_vars] = scaler.transform(bike_test[no_of_variables])
bike_test.describe()
```

### Deviding test data into x and y and add constant

```python
y_bike_test = bike_test.pop('cnt')
X_bike_test = bike_test
# Adding constant variable to test dataframe
X_bike_test = sm.add_constant(X_bike_test)
# predicting using values used by the final model
bike_test_col = X_bike_train_lm.columns
X_bike_test=X_bike_test[bike_test_col[1:]]
# Adding constant variable to test dataframe
X_bike_test = sm.add_constant(X_bike_test)

X_bike_test.info()
```

```python
y_bike_predict = lm.predict(X_bike_test)
from sklearn.metrics import r2_score
r2_score(y_bike_test, y_bike_predict)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_bike_test, y_bike_predict)
mse
```

```python
fig = plt.figure()
plt.scatter(y_bike_test, y_bike_predict)
fig.suptitle('Test vs Predict', fontsize = 15)          
plt.xlabel('y_test', fontsize = 15)                          
plt.ylabel('y_predict', fontsize = 15)     
```
